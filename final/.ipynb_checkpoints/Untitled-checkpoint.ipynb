```python
for l in model.variables:
    print(l.name)
    if(l.name=="conv2d_6/kernel:0"):
        tenL=l
    if(l.name=="conv2d_6/bias:0"):
        tenLb=l
    if(l.name=="conv2d_7/kernel:0"):
        tenL2=l
    if(l.name=="conv2d_7/bias:0"):
        tenLb2=l
    if(l.name=="dense_2/kernel:0"):
        tenLb3k=l
        #print("tenLb3k",tenLb3k)
    if(l.name=="dense_2/bias:0"):
        tenLb3b=l
#tenLb3k2, tenLb3k1= tf.split(tenLb3k,[37*75*128,38*75*128], axis=0)
one, temp = tf.split(tenLb3k,[25*75*128,50*75*128], axis=0)
two, three = tf.split(temp,[25*75*128,25*75*128], axis=0)
```

```python
import time
sunflower_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/592px-Red_sunflower.jpg"
sunflower_path = tf.keras.utils.get_file('Red_sunflower', origin=sunflower_url)
start_time = time.time()
img = tf.keras.preprocessing.image.load_img(
    sunflower_path , target_size=(300, 300))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch
start_time = time.time()
predictions = model.predict(img_array)
print(predictions)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
c=tf.constant(([0.00392156862745098]))
img_array=tf.math.multiply(
    img_array, c, name=None
)
```

```python
conv1=tf.keras.backend.conv2d(
    img_array, tenL, strides=(1, 1), padding='same', data_format=None, dilation_rate=(1,1)
)
bias1=tf.keras.backend.bias_add(conv1, tenLb)
relu1=tf.keras.backend.relu(
    bias1, alpha=0.0, max_value=None, threshold=0
)
pool1=tf.keras.backend.pool2d(relu1, pool_size=(2, 2), strides=(2, 2), padding='valid',data_format='channels_last')

conv2=tf.keras.backend.conv2d(
    pool1, tenL2, strides=(1, 1), padding='same', data_format=None, dilation_rate=(1,1)
)
bias2=tf.keras.backend.bias_add(conv2, tenLb2)
relu2=tf.keras.backend.relu(bias2, alpha=0.0, max_value=None, threshold=0)
pool2=tf.keras.backend.pool2d(relu2, pool_size=(2, 2), strides=(2, 2), padding='valid',data_format='channels_last')
flatten=tf.keras.backend.batch_flatten(pool2)
dense1=tf.keras.backend.dot(flatten,tenLb3k)
dbias1=tf.keras.backend.bias_add(dense1,tenLb3b)
```
