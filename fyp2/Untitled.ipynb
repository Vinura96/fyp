```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
print(tf.__version__)
```

```python
model = tf.keras.models.load_model('con2_300.h5')
```

```python
for l in model.variables:
    print(l.name)
    if(l.name=="conv2d_6/kernel:0"):
        tenL=l
    if(l.name=="conv2d_6/bias:0"):
        tenLb=l
    if(l.name=="conv2d_7/kernel:0"):
        tenL2=l
    if(l.name=="conv2d_7/bias:0"):
        tenLb2=l
    if(l.name=="dense_2/kernel:0"):
        tenLb3k=l
    if(l.name=="dense_2/bias:0"):
        tenLb3b=l
tenLb3k2, tenLb3k1= tf.split(tenLb3k,[37*75*128,38*75*128], axis=0)
```

```python
import time
sunflower_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/592px-Red_sunflower.jpg"
sunflower_path = tf.keras.utils.get_file('Red_sunflower', origin=sunflower_url)
start_time = time.time()
img = tf.keras.preprocessing.image.load_img(
    sunflower_path , target_size=(300, 300))
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch
start_time = time.time()
predictions = model.predict(img_array)
print(predictions)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
c=tf.constant(([0.00392156862745098]))
img_array=tf.math.multiply(
    img_array, c, name=None
)
```

```python
xx=tf.keras.backend.conv2d(
    img_array, tenL, strides=(1, 1), padding='same', data_format=None, dilation_rate=(1,
    1)
)
xx=tf.keras.backend.bias_add(xx, tenLb)
xx=tf.keras.backend.relu(
    xx, alpha=0.0, max_value=None, threshold=0
)
xx=tf.keras.backend.pool2d(xx, pool_size=(2, 2), strides=(2, 2), padding='valid',data_format='channels_last')
############################################

xx=tf.keras.backend.conv2d(
    xx, tenL2, strides=(1, 1), padding='same', data_format=None, dilation_rate=(1,
    1)
)
xx=tf.keras.backend.bias_add(xx, tenLb2)
xx=tf.keras.backend.relu(xx, alpha=0.0, max_value=None, threshold=0)
xx=tf.keras.backend.pool2d(xx, pool_size=(2, 2), strides=(2, 2), padding='valid',data_format='channels_last')
xx=tf.keras.backend.batch_flatten(xx)
xx=tf.keras.backend.dot(xx,tenLb3k)
xx=tf.keras.backend.bias_add(xx,tenLb3b)

# xx=tf.keras.backend.bias_add(xx, tenLb)
# xx=tf.keras.backend.relu(
#     xx, alpha=0.0, max_value=None, threshold=0
# )
sf0, sf1= tf.split(img_array,[152,148], axis=1)
ss0, ss1= tf.split(img_array,[144,156], axis=1)
```
